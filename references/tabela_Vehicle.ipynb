```python
import pandas as pd
from IPython.display import display 

# Lê o conjunto de dados
data = pd.read_csv('../data/vehicle_price_prediction.csv', sep=',', quotechar='"')

# Número de linhas e colunas
num_linhas, num_colunas = data.shape

# Verifica dados faltantes
faltantes = data.isnull().sum()

# Define as variáveis que você quer descrever
variaveis = [
    "make", "model", "year", "mileage", "engine_hp", "transmission",
    "fuel_type", "drivetrain", "body_type", "exterior_color", "interior_color",
    "owner_count", "mileage_per_year", "seller_type", "accident_history"
]



# Criação da tabela descritiva
tabela_descritiva = pd.DataFrame({
    "Nome da variável": variaveis,
    "Tipo": [
        "Categórico nominal",  # make
        "Categórico nominal",  # model
        "Numérico (inteiro)",    # year
        "Numérico (inteiro)",      # mileage
        "Numérico (inteiro)",    # engine_hp
        "Categórico nominal",  # transmission
        "Categórico nominal",  # fuel_type
        "Categórico nominal",  # drivetrain
        "Categórico nominal",  # body_type
        "Categórico nominal",  # exterior_color
        "Categórico nominal",  # interior_color
        "Numérico (inteiro)",    # owner_count
        "Numérico (float)",      # mileage_per_year
        "Categórico nominal",  # seller_type
        "Categórico Ordinal",  # accident_history
    ],
    "Descrição": [
        "Fabricante do veículo",
        "Modelo do veículo",
        "Ano de lançamento do veículo",
        "Quilometragem do veículo",
        "Potência do motor (em HP)",
        "Tipo de transmissão",
        "Tipo de combustível",
        "Tipo de tração",
        "Tipo de carroceria",
        "Cor externa",
        "Cor interna",
        "Número de proprietários anteriores",
        "Quilometragem média por ano",
        "Tipo de vendedor",
        "Histórico de acidentes"
    ],
    "Unidades": [
        "-",      # make
        "-",      # model
        "ano",    # year
        "km",     # mileage
        "HP",     # engine_hp
        "-",      # transmission
        "-",      # fuel_type
        "-",      # drivetrain
        "-",      # body_type
        "-",      # exterior_color
        "-",      # interior_color
        "unidades",  # owner_count
        "km/ano", # mileage_per_year
        "-",      # seller_type
        data['accident_history'].unique(),      # accident_history
    ],
    "Valores Falantes": [faltantes.get(var, 0) for var in variaveis]
})

# Exibe resultados gerais
print(f"Total de amostras (linhas): {num_linhas}")
print(f"Total de atributos (colunas): {num_colunas}\n")

# Exibe a tabela final
display(tabela_descritiva)
```
