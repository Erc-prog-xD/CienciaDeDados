```python
import pandas as pd
from IPython.display import display 

# Lê o conjunto de dados
data = pd.read_csv('../data/world_best_movies.csv', sep=',', quotechar='"')

# Número de linhas e colunas
num_linhas, num_colunas = data.shape

# Verifica dados faltantes
faltantes = data.isna().sum()

# Define as variáveis que você quer descrever
variaveis = [
    "title", "link", "year", "duration", "rating_mpa", "rating_imdb",
    "vote", "budget", "genre", "language", "director",
    "writer", "filming_location", "win", "oscar"
]

# Criação da tabela descritiva
tabela_descritiva = pd.DataFrame({
    "Nome da variável": variaveis,
    "Tipo": [
        "Textual",               # title
        "Textual",               # link
        "Numérico (inteiro)",    # year
        "Data/Hora",             # duration
        "Categórico",            # rating_mpa
        "Numérico (float)",      # rating_imdb
        "Numérico (inteiro)",    # vote
        "Numérico (float)",      # budget
        "Categórico",            # genre
        "Categórico",            # language
        "Textual",            # director
        "Textual",            # writer
        "Categórico",            # filming_location
        "Numérico (inteiro)",            # win
        "Numérico (inteiro)"             # oscar
    ],
    "Descrição": [
        "Título do filme",
        "Link para a página do filme no IMDb",
        "Ano de lançamento do filme",
        "Duração total do filme (horas e minutos)",
        "Classificação indicativa MPA (Motion Picture Association)",
        "Nota média atribuída pelos usuários no IMDb",
        "Número total de votos recebidos no IMDb",
        "Orçamento estimado do filme em dólares",
        "Gênero ou categoria cinematográfica principal",
        "Idioma principal do filme",
        "Diretor principal do filme",
        "Roteirista principal ou equipe responsável pelo roteiro",
        "Local de filmagem principal",
        "Quantidade de prêmios vencidos",
        "Quantidade de prêmios do Oscar recebidos"
    ],
    "Unidades": [
        "-", "-", "ano", "horas e minutos", "-", "0–10", "votos", "USD", "-", "-", "-", "-", "-", "premiação", "premiação"
    ],
    "Valores Falantes": [faltantes.get(var, 0) for var in variaveis]
})

# Exibe resultados gerais
print(f"Total de amostras (linhas): {num_linhas}")
print(f"Total de atributos (colunas): {num_colunas}\n")

# Exibe a tabela final
display(tabela_descritiva)
```
